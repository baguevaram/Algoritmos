```python
import numpy as np
import matplotlib.pyplot as plt

npermut = 0
total = 0
cases = np.zeros(1001)
prob = np.zeros(1001)
minCase = 1000
maxCase = 0


def isortSteps(a):
    v = []
    for i in range(len(a)):
        v.append(a[i])
        
    steps = 0
    for i in range(1,len(v)):
        x = v[i]
        j = i-1
        while (j > -1) and (v[j] > x):
            v[j+1] = v[j]
            j = j -1
            steps = steps + 3
        steps = steps + 1
        v[j+1] = x
        steps = steps + 4
    steps = steps + 1
    return steps

def swap(v,i,j):
    t = v[i]
    v[i] = v[j]
    v[j] = t

def perm(v,i,n):
    global npermut
    global total
    global cases
    
    j=0
    k=0
    l=0
    if i == n:
        npermut = npermut + 1
        k = isortSteps(v)
        total = total + k
        cases[k] = cases[k] + 1
    else:
        for j in range (i,n):
            swap (v, i, j)
            perm (v, i+1, n)
            swap (v, i, j)

n=8
v=[]
for i in range(n):
    v.append(i+1)
    
perm(v,0,len(v))

for k in range(1001):
    prob[k] = cases[k]/float(npermut)
    if (cases[k]!=0) & (minCase > k):
        minCase = k
    if (cases[k]!=0) & (maxCase < k):
        maxCase = k
#for para imprimir las probabilidades como en los Applets
#for k in range(minCase,maxCase+1):
    #print 'prob(',k,')' ,prob[k]
        
plt.bar(range(minCase,maxCase+1),prob[minCase:maxCase+1])
plt.xlabel('Lineas')
plt.ylabel('Probabilidad')
plt.title('Frecuencia del numero de lineas ejecutadas')
print 'Promedio de lineas ejecutadas',(float(total)/npermut)
```
